# Funciones de probabilidad de una variable aleatoria discreta

* Es una función que asigna probabilidades a los valores (x) de una variable aleatoria X

# Función de distribución acumulativa de una variable aleatoria discreta

* Especifica la probabilidad de que una variable aleatoria sea menor o igual que un valor dado



```
F(X) = P(X<=x)
```



**Ejemplo:**

* Sea X(Variable aleatoria discreta) el número de productos vendidos en un almacén. Encontrar la función de distribución acumulada a partir de su función de probabilidad.

````python
import pandas as pd
x = [0,1,2]
f_x = [0.5,0.3,0.2]
print(f"suma probabilidades: {sum(f_x)}")
tabla = pd.DataFrame(list(zip(x,f_x)), columns = ["x", "f(x)"])
tabla
````

# Encontrando la función de distribución acumulativa F(x)



```
F(0) = P(X<=0) = P(X=0) = f(0) = 0.5
F(1) = P(X<=1) = P(X=0 o 1) = f(0) + f(1) = 0.5 + 0.3 = 0.8
F(2) = P(X<=2) = P(X=0, 1 o 2) = f(0) + f(1) + f(2) = 0.5 + 0.3 + 0.2 = 1
```



````python
#Creando la función de distribución acumulativa
F_x = []
for i in range(len(f_x)):
  if i==0:
    F_x.append(f_x[i])
  else:
    F_x.append(f_x[i]+F_x[i-1])
````

````python
#Agregándola al dataframe
tabla["F(x)"] = F_x
tabla.plot(x="x", y="F(x)")
````

# **Distribución de probabilidad de la variable aleatoria binomial**

mide el número de éxitos probables en n experimentos independientes, como sucede con el lanzamiento de dados o monedas. Es un modelo adecuado para procesos donde un detemrinado ejercicio se repite hasta obtener un número establecido de resultados favorables (por primera vez). 

image.png



# Condiciones Experimento binomial

* El experimento consta de una secuencia de n ensayos idénticos
* En cada ensayo hay dos resultados posibles. A uno de ellos se le llama éxito y al otro, fracaso
* La probabilidad de éxito es constante de un ensayo a otro, nunca cambia ys e denota por p. Por ello, la probabilidad de fracaso será 1-p
* Los ensayos son independientes, de modo que el resultado de cualquiera de ellos no influye en el resultado de cualquier otro ensayo


--> en lo que se enfoca la binomial: sea p la probabilidad de éxito y 1-p la probabilidad de fracaso en un solo ensayo, **entonces la probabilidad de obtener x éxitos en n ensayos, está dada por la función** -->

f(x) para x como valores enteros --> variable aleatoria discreta

# Distribución geométrica

* Esta distribución nos permite obtener la probabilidad de un número determinado de repeticiones (x), antes de que se pueda tener éxito la primera vez.
* Número de ensayos requeridos para obtener el primer éxito
1. P(X=x) = P(1-p)**(x-1) --> x = 0,1,2,...

* Número de fallos iguales a x antes de lograr éxito
2. P(y=x) = P(1-p)**(x) --> x = 0,1,2

* **Ejemplo:** Suponga que cada una de las llamadas a una estación de radio tiene una probabilidad de 0.02 de ser respondida. Asumiendo que las llamadas son independientes:
- Cuál es la probabilidad de que le respondan a la décima llamada?

--> lo que se hace es un ensayo de Bernoulli repetidas veces hasta obtener un éxito

image.png

````python
#x: número de llamadas a la estación hasta ser atendido
#éxito: llamada respondida
#fallo: llamada no respondida

p = 0.02 #prob. de éxito -> 2%
q = 1-p #prob. fallo

#respondiendo a la pregunta
x = 10
prob = (q**(x-1))*p
print(prob)
````

````python
#con la librería: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.geom.html
from scipy.stats import geom

prob = geom.pmf(10,0.02)
print(prob)
````

# Distribución Poisson

* Esta distribución describe la probabilidad de que ocurran un número de éxitos en un intervalo definido. El intervalo puede ser de tiempo, distancia, área, volumen o alguna unidad similar

Por ejemplo:
* X = número de clientes que ingresan a un banco en 1 hora
* X = número de accidentes en un tramo de autopista en 1 día
* X = número de bacterias presentes en un litro de agua

En el contexto del modelo:
* x: variable aleatoria discreta (enteros positivos, 0,1,2,... -> variables que se puedan contar)
* lambda: tasa de éxitos en un intervalo de tiempo o espacio
--> se toma una muestra significativa por intervalos (tasa: suma de ingresos/total)
--> ejemplo: 1.4 clientes por minuto

* **Supuestos**
  * La variable aleatoria describe la ocurrencia de un evento en un intervalo definido
  * La probabilidad de ocurrencia es la misma para dos intervalos de igual longitud cualquiera (mu -> lambda -> tasa -> ritmo promedio)
  * La ocurrencia o no ocurrencia en cualquier intervalo es independiente de la ocurrencia o no ocurrencia en cualquier otro intervalo -> las ocurrencias son independientes
  * Dos eventos no pueden ocurrir al mismo tiempo


* Ejemplo: los clientes ingresan a una tienda según una tasa de 0.4 clientes por minuto

  - Calcule la probabilidad de que, en un intervalo de 1 minuto, no ingresen clientes. 
  - Por qué Poisson? --> hay una tasa de éxito en un lapso de tiempo

image.png

````python
#Manual
import math

#x: variable aleatoria (número de clientes que ingresan a la tienda en 1 minuto)
x = 0 #P(X=0) --> Cuál es la probabilidad de que la cantidad de clientes ingresados sea 0?. Mi variable aleatoria (X) es cant. personas ingresas por tasa.
lam = 0.4 #tasa promedio 
prob_manual = (math.exp(-lam)*(lam**x))/math.factorial(x)
print(prob_manual)
````

# Distribución hipergeométrica

* Se usa cuando se hacen ensayos/pruebas sin reemplazo.
--> los ensayos no son independientes (no son de Bernoulli)
--> se usa hipergeométrica

image.png

* **Ejemplo**: Bolsa con 10 pelotas con 1s(5) y 0s(5)

  * x: variable aleatoria --> veces que se obtiene un 0 al sacar 3 números de la bolsa
  * p = 5/10 (valor inicial que va cambiando a medida que van cambiando las pelotas en la bolsa

* De manera general:
  * N = número total de la población
  * k = número de éxitos de la población
  * n = tamaño de la muestra
  * x = número de éxitos de la muestra
------

* K = éxitos posibles --> 5 0s inicial
* N = total eventos al inicio --> 10
* n = muestra --> 3 (cuántas pelotas va a sacar)

* **Problema**: Cuál es la probabilidad de sacar dos 0s al sacar 3 pelotas

````python
#con librería
from scipy.stats import hypergeom
x = 2 #variable aleatoria
N = 10 #total eventos
k = 5 #éxitos posibles
n = 3 #muestra (cuántas pelotas va a sacar)
prob = hypergeom.pmf(x,N,k,n)
print(prob)
````

````python
#manual
import math

def nCr(n,k):
  f = math.factorial
  return f(n) // f(k) // f(n-k)

def g(x,N,n,k): 
  return (nCr(k,x)*nCr(N-k,n-x))/nCr(N,n)

print(g(2,10,3,5))
````
